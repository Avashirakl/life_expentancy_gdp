```python
import pandas as pd

df = pd.read_csv('C:/Users/User/Desktop/life_expentancy_gdp/all_data.csv')
df
```

```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

vvp = []
leb = []
for i in range(0, df.shape[0]):
    vvp.append(int(df['GDP'].values[i]/1000000000))
    leb.append(df['Life expectancy at birth (years)'].values[i])

plt.figure(figsize=(25,5))
sns.barplot(x = vvp, y = leb)
plt.show()
```

```python
life = df['Life expectancy at birth (years)']
year = df['Year']

increased = []
decreased = []
ifincreased = 0

for i in range(df.shape[0]):
    if(life.values[i] - life.values[i-1] and  year.values[i] - year.values[i-1] > 0):
        increased.append("increased")
        ifincreased = int(life.values[i] - life.values[i-1])
        
        if(ifincreased != 0):
            print(f"if increased {ifincreased}")
            
    else:
        decreased.append("decreased")
        
plt.figure(figsize= (10, 5))
plt.bar('L.E. than last year is increased', len(increased), color = 'green')
# plt.bar('L.E. than last year is stay', len(stay), color = 'yellow')
plt.bar('L.E. than last year is decreased', len(decreased), color = 'grey')

# print(f"if increased {ifincreased}")
```

```python
year = []; yearc = []; yearg = []; yearm = []; yeara = []; yearz = []
gdp = []; gdpc = []; gdpg = []; gdpm = []; gdpa = []; gdpz = [];


for i in range(df.shape[0]):
    
    if(df['Country'].values[i] == "Chile"):
        year.append(df['Year'].values[i])
        gdp.append(int(df['GDP'].values[i]/1000000000))
          
    if(df['Country'].values[i] == "China"):
        yearc.append(df['Year'].values[i])
        gdpc.append(int(df['GDP'].values[i]/1000000000))
    
    if(df['Country'].values[i] == "Germany"):
        yearg.append(df['Year'].values[i])
        gdpg.append(int(df['GDP'].values[i]/1000000000))
    
    if(df['Country'].values[i] == "Mexico"):
        yearm.append(df['Year'].values[i])
        gdpm.append(int(df['GDP'].values[i]/1000000000))
        
    if(df['Country'].values[i] == "United States of America"):
        yeara.append(df['Year'].values[i])
        gdpa.append(int(df['GDP'].values[i]/1000000000))
        
    if(df['Country'].values[i] == "Zimbabwe"):
        yearz.append(df['Year'].values[i])
        gdpz.append(int(df['GDP'].values[i]/1000000000))


plt.figure(figsize=(25,25))

plt.subplot(6,2,1)
plt.title("Chile")
plt.plot(year, gdp)

plt.subplot(6,2,2)
plt.title("China")
plt.plot(yearc, gdpc)

plt.subplot(6,2,3)
plt.title("Germany")
plt.plot(yearc, gdpc)

plt.subplot(6,2,4)
plt.title("Mexico")
plt.plot(yearm, gdpm)

plt.subplot(6,2,5)
plt.title("United States of America")
plt.plot(yeara, gdpa)

plt.subplot(6,2,6)
plt.title("Zimbabwe")
plt.plot(yearz, gdpz)
```
